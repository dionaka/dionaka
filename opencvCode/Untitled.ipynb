##读取视频

```python
video=cv2.VideoCapture("ge.mp4")
```

```python
 if not video.isOpened():
    print("error")
    exit() 
```

```python
while True:
    ret, frame=video.read()
    if not ret:
        break
    gray_frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("按q退出",gray_frame)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```

```python
top,bottom,left,right=(50,50,50,50)
#复制法，复制边缘像素
a1=cv2.copyMakeBorder(bb,top,bottom,left,right,cv2.BORDER_REPLICATE)
#反射法，edcba|abcdef|fedcb
a2=cv2.copyMakeBorder(bb,top,bottom,left,right,cv2.BORDER_REFLECT)
#反射法，edcb|abcdef|edcb
a3=cv2.copyMakeBorder(bb,top,bottom,left,right,cv2.BORDER_REFLECT_101)
#外包装法
a4=cv2.copyMakeBorder(bb,top,bottom,left,right,cv2.BORDER_WRAP)
#常量法 
a5=cv2.copyMakeBorder(bb,top,bottom,left,right,cv2.BORDER_CONSTANT,value=210)
```

```python
cv2.imshow("gray",a1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

图像处理
threshold：阈值
1.cv2.THRESH_BINARY   (_INV")//超过阈值取最大值，否则取0(反转）
2.cv2.THRESH_TRUNC//截断
3.THRESH_TOZERO//超过阈值不变，否则取0（反转）

```python
nu2=cv2.cvtColor(nu2,cv2.COLOR_BGR2GRAY)
ret,shuchu=cv2.threshold(nu2,127,255,cv2.THRESH_BINARY)
ret,shuchu2=cv2.threshold(nu2,127,255,cv2.THRESH_BINARY_INV)
ret,shuchu3=cv2.threshold(nu2,127,255,cv2.THRESH_TRUNC)
ret,shuchu4=cv2.threshold(nu2,127,255,cv2.THRESH_TOZERO)
```

```python
du(nu2)
```
